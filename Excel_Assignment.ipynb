# Excel Basics

Q1. **Explain the difference between Relative , Absolute and Mixed Cell Referencing.**

- Relative Cell Reference (A1)

1.Changes automatically when a formula is copied.

2.Used when the formula logic should adjust row/column positions.

- Absolute Cell Reference ($A$1)

1.Does not change when copied.

2.Used for fixed values like tax rate, discount, constants.

- Mixed Cell Reference ($A1 or A$1)

1.Either row or column is fixed.

2.Commonly used in multiplication tables and matrix formulas.
