# Data Cleaning & Functions

Q1. What is data cleaning, and why is it important in data analysis?

*   What are the potential consequences of analyzing unclean or messy data?

*   Explain the common steps involved in cleaning and organizing data.

Ans- Data cleaning is the process of identifying and correcting errors, inconsistencies, and inaccuracies in a dataset.

**Why it is important:**

1.   Ensures accuracy of analysis

2. Improves data quality and reliability

3. Helps in making correct business decisions

4. Reduces errors in formulas, charts, and reports

5. Makes data easier to understand and analyze

**Potential consequences of analyzing unclean or messy data**

1.   Incorrect results and conclusions

2. Misleading trends and insights

3. Poor decision-making

4. Formula errors and calculation issues

5. Loss of credibility and trust in reports

**Common steps involved in cleaning and organizing data**

1. Remove duplicate records

2. Fix spelling and formatting errors

3. Handle missing or blank values

4. Standardize data formats (dates, text, numbers)

5. Remove unnecessary spaces using TRIM

6. Validate data entries

7. Sort and organize data logically






Q2. **How would you sort the following dataset first by "Department" (A-Z) and then by "Salary" (Largest to
Smallest)? Write a step-by-step approach.**

Ans- **Step-by-step approach:**

1. Select the entire dataset

2. Go to Data → Sort

3. In Sort by, choose Department

4. Select A to Z

5. Click Add Level

6. In the second level, choose Salary

7. Select Largest to Smallest

8. Click OK

**Q6. Understand and apply fundamental text functions like LEFT, RIGHT, MID, and LEN.**

1. Extract the first 5 characters from the string "ExcelTipsAreGreat" using the LEFT function.

Ans- =LEFT("ExcelTipsAreGreat",5)

2.Extract the last 4 characters from "DataAnalysis.xlsx" using the RIGHT function.

Ans- =RIGHT("DataAnalysis.xlsx",4)

3. Extract the substring "Tips" from "ExcelTipsAreGreat" using the MID function.

Ans- =MID("ExcelTipsAreGreat",6,4)

4. Count the total number of characters in the string "Hello World!" using the LEN function

Ans- =LEN("Hello World!")

5. Create a formula to extract the middle 6 characters from "12345-67890-ABCDE".

Ans- =MID("12345-67890-ABCDE",7,6)

**8.Understanding TODAY() and NOW()**

1.What is the difference between TODAY() and NOW() in Excel? Provide an example of when you would use each function.

Ans- **TODAY()**

a. Use when you only need the current date.

b. Common uses: deadlines, age calculation, due dates, daily reports.

Practical example:
Calculate days left until a project deadline: =Deadline_Date - TODAY()

  **NOW()**

Use when you need both the current date and current time.

Common uses: time tracking, logging entry times, attendance records.

Practical example:
Record the exact time a form was submitted or a task was completed.

2. If cell A1 contains the date 2025-06-10, write a formula using TODAY() to determine how many days are left until the date.

Ans- A1 - TODAY()

3. Write an Excel formula using NOW() to display the current date and time in the format MM/DD/YYYY HH:MM AM/PM.

Ans- TEXT(NOW(),"MM/DD/YYYY HH:MM AM/PM")

4. If a cell contains =TODAY(), what will happen when the worksheet is reopened the next day? Explain

Ans- If a cell contains =TODAY(), then when the worksheet is reopened the next day, the value in that cell will automatically update to the new current date.
